```python
import pandas as pd
from pprint import pprint

tickets = pd.read_csv("data_safe_copy.csv")

print(tickets.head())

tickets.info()
```

```python
tickets_data = pd.read_csv("data_safe_copy.csv", parse_dates=["opened_at"])
tickets_data["category"] = tickets_data["category"].str.strip().str.lower()
unique_values_cat = tickets_data["category"].unique()
unique_values_pri = tickets_data["priority"].unique()
print(unique_values_cat, unique_values_pri)
```

```python
resolution_minutes = tickets_data["resolution_minutes"] = pd.to_numeric(tickets_data["resolution_minutes"], errors="coerce")

resolution_minutes_check = resolution_minutes.isna().sum()

resolution_minutes_len_dropnan = len(resolution_minutes.dropna()) #resolution_minutes.dropna() drops any values with NaN in "resolution_minutes" in resolution_minutes column

resolution_minutes_dropnan = resolution_minutes.dropna()

#print(resolution_minutes_check) #100
#print(resolution_minutes_dropnan) #400

if (len(tickets) - resolution_minutes_len_dropnan) == resolution_minutes_check:
    print("True")
else:
    print("False")

tickets_clean = tickets_data.dropna(subset=["resolution_minutes"])

verify_non_negative = (tickets_clean["resolution_minutes"] <= 0).sum()
if verify_non_negative == 0:
    print("0")

verify_normalized_data = tickets_clean["category"].unique()
print(verify_normalized_data)

print(tickets.shape) #500 - before cleaning
print(tickets_clean.shape) #400 - after cleaning
```

```python
summary_dictionary = {
    "total_records" : len(tickets),
    "clean_records" : len(tickets_clean),
    "missing_resolution" : int(resolution_minutes_check),
    "avg_resolution" : float(tickets_clean["resolution_minutes"].mean()),
    "max_resolution" : tickets_clean["resolution_minutes"].max()
}

if summary_dictionary["clean_records"] + summary_dictionary["missing_resolution"] == summary_dictionary["total_records"]:
    print("all's good")
else:
    print("all's not good")

print(" ")
pprint(summary_dictionary)
```
